Juliaは電卓のように使えるので、適当に計算をして、shift+enterキーで実行することができます。
四則演算：

```julia
1+3
```

```julia
sin(pi)
```

```julia
sin(0.1)^2+cos(0.1)^2
```

のように実行できます。なお、円周率はpiで出せます。

Juliaはベクトルや行列の扱いにも強く、ベクトルは

```julia
x = [1,2,3]
```

や

```julia
x = [1
    2
    3]
```

のように定義できます。また、等間隔で並んだ点が欲しい場合には、

```julia
x = range(0,2pi,100)
```

とすれば、100個の点が得られます。なお、2piのように、記号と数字をくっつけることもできます。
また、関数は

```julia
f(x) = x^2
```

のように定義できます。そして、xが複数の数字の集合（例えばベクトルとか行列とか）の場合、それぞれの要素に関数を作用させることもでき、その場合は

```julia
y = f.(x)
```

のように.をつけることでまとめて計算ができます。

## プロット
次はプロットしてみます。まず、Plotsというパッケージをインストールします。パッケージのインストールをJupyter notebook上で行うには

```julia
import Pkg
```

とします。Plotsを入れる場合には、

```julia
Pkg.add("Plots")
```

とします。そして、

```julia
plot(x,cos.(x))
```

とすると、グラフが書けます。もし、さらにこの図にグラフに追加したい場合には、

```julia
plot!(x,sin.(x))
```

のようにplot!と!をつけた関数を使います。今、凡例はy1やy2となっていますが、これは、

```julia
plot(x,cos.(2x),label="cos")
plot!(x,sin.(2x),label="sin")
```

のように、labelというものを使うことで色々描けます。細かなオプションは、[こちら](https://docs.juliaplots.org/latest/tutorial/)を見ればわかるかもしれません。
